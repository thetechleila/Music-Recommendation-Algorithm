## Univariate Analysis
___

> ### Pre-Exisiting Info About the *"release_date"* Column
>
>> The year the song was released expressed as a 4 digit integer.
___

```python
#Inspect "release date" column by getting count of unique values (years)
df["release_date"].value_counts(ascending=False)

#It seems as though the amount of releases may increase every year from 1953 to 2017
```

**The dataset spans 70 years, 1950 - 2020**

```python
#Group the release dates and song counts by decade using floor division for understandability
release_count_dec = df["release_date"]//10*10

#Get count of songs grouped by decade
release_count_dec.value_counts()
```

```python
#Visualize the distribution of songs from 1950-2020 (in increments of 10/decades)

sns.histplot(data = df, x = "release_date", color="pink", edgecolor ='black', linewidth= 1)
plt.title("Distrubtion of Songs Released by Year (1950-2019)")
plt.xlabel("Song Release Year")
plt.ylabel("Amount of Songs")
plt.show()
```

```python
#Visualization the count of songs, grouped by decade specifically
sns.histplot(data = release_count_dec, color="green", edgecolor ='black', linewidth= 1)
plt.title("Count of Songs in the Dataset Released by Decade 1950-2010")
plt.xlabel("Dataset Song Release Decade")
plt.ylabel("Song Count")
plt.show()
```



> ### *Insights* from the Univariate Analysis of the "release_date" Column

> ### Pre-Exisiting Info About the *"age"* Column
>
>>  A score from 0 to 1 expressing how “old” a song is from our perspective. 1 being the oldest, and 0 being the newest expressed as a single integer.
___

```python
#Get overall count of songs considered old or new, where 1 = the oldest  and 0 = the newest
df["age"].value_counts(ascending=False)
```

**70 unique age integer values contained in the dataset ranging from 0 to 1.** 

```python
#Specifically find count of the youngest songs where age = 0

youngest_songs = df [df["age"]==0]

youngest_songs = youngest_songs.value_counts().sum()

print ("The total count of the absolute youngest songs in the dataset is:", youngest_songs, "\n  There aren't any songs in the dataset with an age of true 0.")
```

```python
#Specifically find count of the oldest songs where age = 1

oldest_songs = df[df["age"]== 1]

oldest_songs = oldest_songs.value_counts().sum()

print ("The total count of the absolute oldest songs in the dataset is:", oldest_songs)
```

```python
#Get count of songs older than 0.5

over_05 = df[df["age"] > 0.5]

over_05 = over_05.value_counts().sum()

print ("There are", over_05, "songs over the age of 0.5 in this dataset.")
```

```python
#Get count of songs younger than 0.2

below_05 = df[df["age"] < 0.5]

below_05 = below_05.value_counts().sum()

print ("There are", below_05, "songs under the age of 0.5 in this dataset.")
```

```python
sns.histplot(data = df, x = "age", color="red", edgecolor ='black', linewidth= 1)
plt.title("Age of Songs in the Dataset from 1950-2019")
plt.xlabel("Age Range of Songs (0= Youngest, 1=Oldest)")
plt.ylabel("Amount of Songs")
plt.show()
```
