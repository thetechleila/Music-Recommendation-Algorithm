## Univariate Analysis
___

> ### Pre-Exisiting Info About the *"release_date"* Column
>
>> The year the song was released expressed as a 4 digit integer.

```python
#Inspect "release date" column by getting count of unique values
release_count = df["release_date"].value_counts(ascending=True)

#It seems as though the amount of releases may increase every year from 1953 to 2017
```

```python
#Visualize the count of songs by year in a histplot
sns.histplot(data = df, x = 'release_date')
plt.title('Number of songs released since 1950')
plt.xlabel('Release Year')
plt.ylabel('Number of songs released')
plt.show()

```

```python
sns.histplot(data = release_count)
plt.title('Number of songs released in each decade since 1950s')
plt.xlabel('Decades')
plt.ylabel('Number of songs released')
plt.show()
```
